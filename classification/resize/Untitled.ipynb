```python
image_list = []
resized_images = []
```

```python
#for filename in glob.glob('C:/Users/avish/Desktop/All/.sliit/SLIIT Y3S1/Design Project - I - EC3061/cnn/try 2/dataset/input/train/soccer/*.jpg'):
for filename in glob.glob('C:/Users/avish/Desktop/All/.sliit/SLIIT Y3S1/Design Project - I - EC3061/cnn/resize/*.jpg'):

    print(filename)
    img = Image.open(filename)
    image_list.append(img)

imagex = matplotlib.image.imread("C:/Users/avish/Desktop/All/.sliit/SLIIT Y3S1/Design Project - I - EC3061/cnn/resize/aa.jpg")
print()
print('before rescaling')
print('size of the image = ',imagex.shape)
plt.imshow(imagex)
```
