```python
rgb_weights = [0.2989, 0.5870, 0.1140]   #NTSC weights
imagex = matplotlib.image.imread("C:/Users/avish/Desktop/All/.sliit/SLIIT Y3S1/Design Project - I - EC3061/data/New folder/tomato1.jpg")
plt.imshow(imagex)
```
